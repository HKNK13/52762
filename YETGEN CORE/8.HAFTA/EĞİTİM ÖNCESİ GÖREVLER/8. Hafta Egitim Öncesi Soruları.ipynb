# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python

a=int(input("Kaç tane ürün gireceksiniz"))
while True:
    if a>=1:   
        ürün_adı=str(input("Ürün ismini giriniz: ")).capitalize()
        ürün_fiyatı=float(input("Ürün fiyatını  giriniz: "))
        with open("urunler.txt","a",encoding="utf-8") as file:
            file.write(f"Ürün ismi :{ürün_adı} ürün fiyatı:{ürün_fiyatı}\n")
        a-=1
    else:
        break 
print("Dosyanız yazıldı.Kontrol ediniz.")
        


    


````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def değiştir(dosya_adı,eski_isim,yeni_isim):
    with open(dosya_adı,"r+",encoding="utf-8") as file:
        text=file.read()
        yeni_text=text.replace(eski_isim,yeni_isim)
        file.seek(0)
        file.write(yeni_text)
        file.truncate()

değiştir("urunler.txt","Karam","Canga")
değiştir("urunler.txt","17","19")
    
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class vehicle:
    def __init__(self,name,speed,mileage):
        self.name=name
        self.speed=speed
        self.mileage=mileage
class bus(vehicle):
    pass
school=bus("Volvo",200,17)
print("Name:",school.name)
print("Speed:",school.speed)
print("Mileage:",school.mileage)
````

# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def vki_hesapla(kilo, boy):
    return kilo / (boy ** 2)

def vki_siniflandir(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 25:
        return "Normal Kilolu"
    elif 25 <= vki < 30:
        return "Fazla Kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "İleri Derecede Obez"

def kullanici_veri_al():
    ad_soyad = input("İsim Soyisim: ")
    yas = int(input("Yaş: "))
    boy = float(input("Boy (metre cinsinden, örn. 1.75): "))
    kilo = float(input("Kilo (kg): "))
    
    vki = vki_hesapla(kilo, boy)
    sinif = vki_siniflandir(vki)
    
    return ad_soyad, yas, boy, kilo, vki, sinif

def verileri_kaydet(dosya_adi, veriler):
    with open(dosya_adi, "a") as dosya:
        for veri in veriler:
            dosya.write(f"{veri[0]}, {veri[1]}, {veri[2]}, {veri[3]}, {veri[4]:.2f}, {veri[5]}\n")

def ana_program():
    kac_giris = int(input("Kaç giriş yapmak istiyorsunuz? "))
    veriler = []
    
    for _ in range(kac_giris):
        veri = kullanici_veri_al()
        veriler.append(veri)
    
    dosya_adi = "vki_verileri.txt"
    verileri_kaydet(dosya_adi, veriler)
    print(f"Tüm veriler {dosya_adi} dosyasına kaydedildi.")
ana_program()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime
menu = {
    "banana": 5.0,
    "strawberry": 7.0,
    "blueberry": 8.0,
    "spinach": 3.0,
    "protein_powder": 10.0,
    "almond_milk": 4.0,
    "honey": 2.0
}
def menu_goster():
    print("----- Smoothie Menüsü -----")
    for urun, fiyat in menu.items():
        print(f"{urun.capitalize()}: {fiyat} TL")
def smoothie_olustur():
    toplam_fiyat = 0
    secilen_urunler = []
    
    while True:
        urun = input("Eklemek istediğiniz ürünü seçin (çıkmak için 'q' tuşlayın): ").lower()
        if urun == 'q':
            break
        if urun in menu:
            secilen_urunler.append(urun)
            toplam_fiyat += menu[urun]
        else:
            print("Geçersiz ürün seçimi, lütfen menüden bir ürün seçin.")

    return secilen_urunler, toplam_fiyat
def satin_al(secilen_urunler, toplam_fiyat):
    tarih = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open("smoothie_satis_log.txt", "a") as dosya:
        dosya.write(f"Tarih: {tarih}, Ürünler: {', '.join(secilen_urunler)}, Toplam Fiyat: {toplam_fiyat} TL\n")
    print(f"Toplam {toplam_fiyat} TL ödemeniz gerekmektedir. Afiyet olsun!")
def ana_program():
    while True:
        print("1. Menüyü görüntüle")
        print("2. Smoothie satın al")
        print("3. Çıkış")
        secim = input("Seçiminiz: ")

        if secim == '1':
            menu_goster()
        elif secim == '2':
            secilen_urunler, toplam_fiyat = smoothie_olustur()
            satin_al(secilen_urunler, toplam_fiyat)
        elif secim == '3':
            print("Çıkış yapılıyor...")
            break
        else:
            print("Geçersiz seçim, lütfen tekrar deneyin.")
ana_program()
```
